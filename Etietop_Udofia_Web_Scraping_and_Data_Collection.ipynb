```python
import requests

# Send an HTTP GET request
url = "https://livescore.com"
response = requests.get(url)

# Check the response status code
print("Response Status Code:", response.status_code)

# Print the first 500 characters of the response content
print("Response Content:")
print(response.text[:500])
```

```python
print(response.content)
```

```python
from bs4 import BeautifulSoup

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Print the prettified HTML
print(soup.prettify())
```

```python
header = soup.find("h1")
print("Header:", header)

# Find all <a> tags and extract their href attributes
links = soup.find_all("a")
for link in links:
    print("Link:", link.get("href"))
```

```python
url1 = "https://pinterest.com"
```

```python
response = requests.get(url1)
```

```python
if response.status_code == 200:
    # HTML parsing using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# extract images from pinterest
```

```python
img_tags = soup.find_all('img')
img_urls = [img.get('https://www.pinterest.com/pin/74309462593356668/') for img in img_tags]
```

```python
download_dir = r'C:\Users\ancep\Desktop\AI6\Images'
if not os.path.exists(download_dir):
    os.makedirs(download_dir)
```

```python
for img_url in img_urls:
        try:
            img_data = requests.get(img_url).content
            img_filename = os.path.join(download_dir, os.path.basename(img_url))
            with open(img_filename, 'wb') as img_file:
                img_file.write(img_data)
            print(f"Downloaded: {img_filename}")
        except Exception as e:
            print(f"Failed to download {img_url}: {str(e)}")
else:
    print(f"Error retrieving content from {url}. Status code: {response.status_code}")
```
